```python
import torch
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Load the dataset from the .pkl file
dataset_path = '../../dataset_42/TranslatingMNIST/TranslatingMNIST-digit=3.pkl'
dataset = torch.load(dataset_path)

data = dataset['data']
targets = dataset['targets']

# Function to visualize a single sequence
def visualize_sequence(data_sequence):
    fig, ax = plt.subplots()
    ax.set_axis_off()

    def update(frame_idx):
        ax.imshow(data_sequence[:, :, frame_idx], cmap='gray')
        return ax

    ani = animation.FuncAnimation(fig, update, frames=data_sequence.shape[-1], interval=100, blit=False)
    plt.show()

# # Visualize a few sequences
# for i in range(5):  # Visualize the first 5 sequences
#     print(f"Visualizing sequence {i+1}")
#     visualize_sequence(data[i, 0])

# If you prefer to save the animations as GIFs:
def save_animation(data_sequence, file_name):
    fig, ax = plt.subplots()
    ax.set_axis_off()

    def update(frame_idx):
        ax.imshow(data_sequence[:, :, frame_idx], cmap='gray')
        return ax

    ani = animation.FuncAnimation(fig, update, frames=data_sequence.shape[-1], interval=100, blit=False)
    ani.save(file_name, writer='imagemagick', fps=10)

# Save a few sequences as GIFs
for i in range(20):
    save_animation(data[i, 0], f'sequence_{i+1}.gif')
```
